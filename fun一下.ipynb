```python
#

import requests as r 
import shutil
import os
from bs4 import BeautifulSoup
from time import sleep
k=1
while True:
    res=r.get('https://fun1shot.com/list/%s?id=6'%k)
    soup=BeautifulSoup(res.text,'html5lib')
    title=[i.text.replace('?','')
                 .replace('\u3000','')
                 .replace('','')
                 .replace('@','')
                 .replace('"','')
                 .replace('.','')
                 .replace(':','')
                 .replace('!','')
                 .replace('/','')
                 .replace('*','')
                 .replace('<','')
                 .replace('>','')
                 .replace('(','')
                 .replace(')','')
                 .replace(' ','')for i in soup.select('div.caption > h5.caption-title')]
    artical_url=['https://fun1shot.com'+i['href'] for i in soup.select('div.row > a')]
    if artical_url==[]:
        break
    i=0
    for url in artical_url:
        path='E:\\\\\\fun\\\\%s'%title[i]
        if os.path.isdir(path):
            i+=1
            continue
        else:
            os.mkdir(path)
        res=r.get(url)
        soup=BeautifulSoup(res.text,'html5lib')
        pic_url=[i['src']for i in soup.select('img.article_img')]
        j=1
        for pic in pic_url:
            print(pic)
            res=r.get(pic,stream=True)
            with open(path+'\\%s.jpg'%j,'wb')as f:
                shutil.copyfileobj(res.raw,f)
            j+=1
        i+=1
        sleep(5)
    k+=1
```
