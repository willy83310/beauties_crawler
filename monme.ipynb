```python
import requests as r
import shutil
import os
import time
from bs4 import BeautifulSoup
req=r.get("http://acg.ad2iction.com/?p=93186")
soup=BeautifulSoup(req.text,"html5lib")
title=soup.select("h1 > span")
pics=soup.select('p > img')
pics=[sou["data-wp-content-image-lazy-src"] for sou in pics]
path="/home/ubuntu/Documents/monme/%s"%title[0].text
if not os.path.isdir(path):
    os.mkdir(title[0].text)
i=0
for url in pics:
    res = r.get(url , stream = True)
    with open(path+'/%s.jpg'%i,'wb') as f:
        shutil.copyfileobj(res.raw,f)
        i+=1
        time.sleep(2)
```

```python
import requests as r
import shutil
import os
from bs4 import BeautifulSoup
k=0
while True:
    html="http://acg.ad2iction.com/?cat=1&paged={}"
    req=r.get(html.format(k))
    if req.status_code==400:
        break
    soup=BeautifulSoup(req.text,'html5lib')
    source=soup.select('article > a')
    title=[sou["title"] for sou in source]
    link=[sou['href'] for sou in source]
    print(link)
    j=0
    for lin in link :
        res=r.get(lin)
        ressoup=BeautifulSoup(res.text,'html5lib')
        try:
            pics=ressoup.select('.alignnone')
            pic=[pic["data-wp-content-image-lazy-src"] for pic in pics]
            pic.pop()
            path="E:\\專題資料\\圖片\\萌咩誌\\%s"%title[j]
            if not os.path.isdir(path):
                os.mkdir(path)
            else:
                continue
            i=1
            for url in pic:
                print(url)
                picture=r.get(url, stream = True)
                with open(path+"/"+'%s.jpg'%i,'wb') as f:
                    shutil.copyfileobj(picture.raw,f)
                i+=1
            j+=1
        except:
            print(lin)
            continue
    k+=1
```
